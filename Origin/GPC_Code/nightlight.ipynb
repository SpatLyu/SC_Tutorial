# 相关系数和其他方法对比

```python
from scipy.stats import pearsonr
import numpy as np
from lingam import DirectLiNGAM
```

```python

x = xMatrix.flatten()
y = yMatrix.flatten()

# delete NaN
mask = ~np.isnan(x) & ~np.isnan(y)
x_clean = x[mask]
y_clean = y[mask]

#  Pearson and p value
r, p_value = pearsonr(x_clean, y_clean)

print(f"Pearson correlation: r = {r:.4f}, p = {p_value:.4e}")



# 假设 xMatrix, yMatrix 是二维数组，样本行数相同
# 将它们合并成一个二维数组，列代表变量
data = np.column_stack((xMatrix.flatten(), yMatrix.flatten()))

# creat model
model = DirectLiNGAM()

# fit
model.fit(data)

# the result of lingam
causal_matrix = model.adjacency_matrix_

print("LiNGAM :")
print(causal_matrix)
```
